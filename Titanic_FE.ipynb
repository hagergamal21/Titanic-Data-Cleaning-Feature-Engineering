## Explore Data

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df_copy['Title'] = df_copy['Name'].str.extract(r' ([A-Za-z]+)\.')
```

```python
df_copy.head()
```

```python
df_copy['Title'] = df_copy['Title'].replace(
    ['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 
     'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare'
)
df_copy['Title'] = df_copy['Title'].replace({'Mlle':'Miss', 'Ms':'Miss', 'Mme':'Mrs'})
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_copy['Title'] = le.fit_transform(df_copy['Title'])
```

```python
df_copy.drop(columns=['Name'], inplace=True)
```

```python
df_copy.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_copy['Sex'] = le.fit_transform(df_copy['Sex'])
```

```python
df_copy['Embarked'].unique()
```

```python
df_copy[df_copy['Embarked'].isnull()]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize encoder
encoder = OneHotEncoder(drop='first', sparse_output=False)
```

```python
# Fit and transform
encoded = encoder.fit_transform(df_copy[['Embarked']])
```

```python
# Put back into DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['Embarked']))
df_copy = pd.concat([df_copy, encoded_df], axis=1).drop(columns=['Embarked'])
```

```python
df_copy.head()
```

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
